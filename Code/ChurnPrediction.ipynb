### One Hot Encoding

```python
# One hot encoding 
categorical_columns = train_df.select_dtypes(include=['object', 'string']).columns
print(categorical_columns)

# List of categorical columns to be converted to dummy variables
categorical_columns = ['SubscriptionType', 'PaymentMethod', 'PaperlessBilling', 'ContentType',
       'MultiDeviceAccess', 'DeviceRegistered', 'GenrePreference', 'Gender',
       'ParentalControl', 'SubtitlesEnabled']

# Create dummy variables for training data
train_df = pd.get_dummies(train_df, columns=categorical_columns, drop_first=True)

# Create dummy variables for test data
test_df = pd.get_dummies(test_df, columns=categorical_columns, drop_first=True)
```

### Filtering Features

```python
# Split the features and target variable for training
X = train_df.drop(columns=['Churn'])
y = train_df['Churn']
```

```python
# Random classifier 
from sklearn.ensemble import RandomForestClassifier
import pandas as pd

# X = features and y = target variable
model = RandomForestClassifier(random_state=42)
model.fit(X, y)

# Get feature importance scores
feature_importances = pd.DataFrame(model.feature_importances_,
                                   index=X.columns,
                                   columns=['importance']).sort_values('importance', ascending=False)

# Add the feature names as a column
feature_importances.reset_index(inplace=True)
feature_importances.rename(columns={'index': 'feature'}, inplace=True)

print(feature_importances)
```

```python
# Feature importance threshold
threshold = 0.01

# Filter features based on importance
important_features = feature_importances[feature_importances['importance'] > threshold]['feature'].tolist()

# Filter the original dataset to keep only important features
X_important = X[important_features]
test_df_important = test_df[important_features]
```

`X_important` is the training dataset after all and `test_df_import` is my test dataset.

### Creating validation Set

```python
# Initialize the scaler
scaler = StandardScaler()

X_important = scaler.fit_transform(X_important)
test_df_important = scaler.transform(test_df_important)
                                   
# Split into training and validation sets
X_train, X_val, y_train, y_val = train_test_split(X_important, y, test_size=0.2, random_state=42)                                  
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score

# Initialize the model
model = LogisticRegression(random_state=42, max_iter=1000, class_weight='balanced')

# Train the model on the training data
model.fit(X_train, y_train)
```

### Grid Search 

```python
from sklearn.model_selection import GridSearchCV

# Define the parameter grid
param_grid = {
    'penalty': ['l1', 'l2', 'elasticnet'],
    'C': [0.01, 0.1, 1, 10, 100],
    'solver': ['lbfgs', 'liblinear', 'saga']
}

# Initialize GridSearchCV
grid_search = GridSearchCV(model, param_grid, cv=5, scoring='roc_auc', n_jobs=-1)

# Fit the model
grid_search.fit(X_train, y_train)

# Best parameters and score
print(f"Best parameters: {grid_search.best_params_}")
print(f"Best score: {grid_search.best_score_}")
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, roc_auc_score

# Get the best model
best_model = grid_search.best_estimator_
best_model.fit(X_train, y_train)


# Predict on the validation set
y_pred = best_model.predict(X_val)
y_pred_proba = best_model.predict_proba(X_val)[:, 1]

# Evaluate the model
print("Confusion Matrix:")
print(confusion_matrix(y_val, y_pred))
print("Classification Report:")
print(classification_report(y_val, y_pred))
print(f"Accuracy: {accuracy_score(y_val, y_pred)}")
print(f"AUC-ROC: {roc_auc_score(y_val, y_pred_proba)}")
```
